### Análise dos foco de calor de 1998 a 2019

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel("focos_calor_1998_2019.xlsx")

# Linhas, colunas
df.shape
```

```python
# Visualizando as 5 primeiros linhas
df.head()
```

```python
# Vamos eliminar as colunas indesejadas
df = df[["Ano", "Mês", "Número", "Período"]]

df.head()
```

```python
# Verificando quantidades e tipos
df.info()
```

### Análise Inicial


```python
# Qual a maior quantidade de focos de calor
df["Número"].max()
```

```python
# Qual mês/ano teve a maior quantidade de focos de calor?
df[df["Número"] == df["Número"].max()]
```

### Eliminando linhas indesejáveis

```python
# Vamos manter todas as linhas onde mês != Total
df = df[df["Mês"] != "Total"]
```

```python
# Qual mês/ano teve a maior quantidade de focos de calor?
df[df["Número"] == df["Número"].max()]
```

```python
# Vamos manter todas as linhas onde ano != "maximo"
df = df[df["Ano"] != "Máximo"]

# Vamos manter todas as linhas onde ano != "média"
df = df[df["Ano"] != "Média"]

# Vamos manter todas as linhas onde Ano != minimo
df = df[df["Ano"] != "Mínimo"]
```

```python
# Qual mês/ano teve a maior quantidade de focos de calor?
df[df["Número"] == df["Número"].max()]
```

### Vamos fazer alguns gráficos



```python
# Selecionar somente os dados de 2007
df_2007 = df[df["Ano"] == 2007]
```

```python
# Gráfico de linha (que é o gráfico padrão)
df_2007.plot(x="Mês", y="Número")
```

```python
# Gráfico de barras
df_2007.plot.bar(x="Mês", y="Número")
```

```python
# Grafico de pizza 
df_2007.plot.pie(x="Mês", y="Número")
```

### Grafico com os focos de 2016, 2017 e 2018

```python
# Importando o matplotlib
import matplotlib.pyplot as plt
```

```python
# Vamos fazer um gráfico por ano, então precisamos selecionar os dados 2016
plt.plot("Mês", "Número", data=df[df["Ano"] == 2016], label="2016")

plt.plot("Mês", "Número", data=df[df["Ano"] == 2017], label="2017")

plt.plot("Mês", "Número", data=df[df["Ano"] == 2018], label="2018")

# Exibir legenda
plt.legend()

# titulo
plt.title("Focos de calor - 2016, 2017 e 2018")

# nome dos eixos
plt.xlabel("Mês")
plt.ylabel("Quantidade")

# melhorando o eixo x
plt.xticks(rotation=45)  # (meses)
```
